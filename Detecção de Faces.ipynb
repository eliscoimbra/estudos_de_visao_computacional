### Bibliotecas

```python
#Open CV
import cv2
import matplotlib.pyplot as plt
```

### Carregamento da Imagem

```python
imagem= cv2.imread('./imagens_deteccao_faces/people1.jpg')
```

```python
#quantidade de pixels de altura e largura
#e quantidade de canais de cores
imagem.shape
```

```python
if imagem is not None:
    cv2.imshow('Minha Imagem', imagem)
    cv2.waitKey(0)  # Espera uma tecla ser pressionada
    cv2.destroyAllWindows()  # Fecha todas as janelas
else:
    print("Erro ao carregar a imagem.")
```

```python
if imagem is not None:
    imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)  # Converte de BGR para RGB
    plt.imshow(imagem_rgb)
    plt.axis('off')
    plt.show()
else:
    print("Erro ao carregar a imagem.")
```

```python
imagem = cv2.resize(imagem, (1000, 600))
if imagem is not None:
    imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)  # Converte de BGR para RGB
    plt.imshow(imagem_rgb)
    plt.axis('off')
    plt.show()
else:
    print("Erro ao carregar a imagem.")
```

```python
imagem = cv2.resize(imagem, (1000, 600))
if imagem is not None:
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)  # Converte de BGR para RGB
    plt.imshow(imagem_rgb, cmap = 'gray')
    plt.axis('off')
    plt.show()
else:
    print("Erro ao carregar a imagem.")
```

### Detecção de Faces

```python
detector_facial = cv2.CascadeClassifier('./imagens_deteccao_faces/haarcascade_frontalface_default.xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
```

```python
deteccoes
```

```python
len(deteccoes)
```

```python
for x, y , w, h in deteccoes:
    print(x,y,w,h)
    cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 255), 2)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Mostrar com matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(imagem_rgb)
plt.axis('off')  
plt.title('Imagem com Detecções')
plt.show()
```
