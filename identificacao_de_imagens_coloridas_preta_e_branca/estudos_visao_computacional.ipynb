## Identificador de Cores das Imagens 

```python
def detectar_cor_da_imagem(caminho_imagem):
    imagem = cv2.imread(caminho_imagem)
    
    if len(imagem.shape) == 3:
        
        ##método que separa os canais de cor(RGB)
        ## o método split separa  em três array referentes aos canais de cor
        ## em azul(blue), verde(green), vermelho(red) respectivamente
        
        b,g, r = cv2.split(imagem)
        
        ##compara as entradas dos arrays de canais de cores
        ### se os array b, g e b, r  iguais logo r, g são iguais
        
        if np.array_equal(b,g) and np.array_equal(b, r):
            valores_canais = np.unique(r)
            
            
            if np.isin(valores_canais, [0, 255]).all() and len(valores_canais)<=2:
                return("Imagem binária em tons de preto e branco o cinza")
            elif np.any((valores_canais >0) & (valores_canais <255)):
                return "Escala de cinza em tons intermediários"
            
        else: 
            return "Imagem Colorida"
```
